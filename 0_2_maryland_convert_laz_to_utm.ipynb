# Break long list of files into folders of 50 files so that they may be downloaded.

```python
import os
import shutil
# Divide files into groups of 200 so they can be downloaded and processed in managably sized chunks.
file_counter = 0
chunk_amount = 50

laz_utm_fp = '/content/drive/MyDrive/crane_maryland/laz/laz_fema_2012_forested_utm/'
laz_utm_fp_dest = '/content/drive/MyDrive/crane_maryland/laz/laz_fema_2012_forested_utm/batch_50/'

for file_name in os.listdir(laz_utm_fp):
    origin_file_path = os.path.join(laz_utm_fp,file_name)
    print(origin_file_path)
    if(os.path.isfile(origin_file_path)):  
        if(file_counter % chunk_amount == 0):
            start_num = (file_counter//chunk_amount)*chunk_amount
            dir_name = "laz_utm_"+ str(start_num)+"-"+str(start_num+(chunk_amount-1))+"/"
            new_dir_path = os.path.join(laz_utm_fp_dest,dir_name)
            print(new_dir_path)
            if not os.path.exists(new_dir_path):
                os.mkdir(new_dir_path)

        shutil.copy(origin_file_path, os.path.join(new_dir_path,file_name))
        print(file_counter, origin_file_path)
        file_counter = file_counter + 1            

    
```

```python
import subprocess
import time

# example command
# gdaldem slope E:\a_new_orgs\carleton\hist5706-maryland\lidar_files\reprojected\20120203_17SPD7277_utm_FILTERED_[1.1_2.1]_RASTER_Z_2021-10-30_16h50_37_125.tif 
# C:/Users/User/AppData/Local/Temp/processing_QIEzpg/6c1743fc49544eba9ae8c25d891f86e8/OUTPUT.tif
# -of GTiff -b 1 -s 1.0


input_raster_fp = "/content/drive/MyDrive/crane_maryland/laz/laz_fema_2012_forested_utm/batch_50/laz_utm_0-49/dem_tifs/"
output_slope_fp = "/content/drive/MyDrive/crane_maryland/laz/laz_fema_2012_forested_utm/batch_50/laz_utm_0-49/dem_tifs/slope_tifs/"


# Set crs
import os
#process .tif files one at a time
for root, dirs, files in os.walk(input_raster_fp, topdown=False):
   for name in files:      
      #print(name)
      name_ext = name[-4:].lower()
      if(name_ext==".tif"):
          tif_in_path = os.path.join(input_raster_fp, name)
          slope_out_path = os.path.join(output_slope_fp, (name[:-4]+"_slope"+name_ext))
        
          command = '"gdaldem" slope "' + tif_in_path + '" "' + slope_out_path + '" -of GTiff -b 1 -s 1.0'
                    
          print("!"+command)
          subprocess.call(command, shell=True)
          # just in case, let the sub process run

          time.sleep(4)
```
